```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# stemming
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
print(cv.get_feature_names_out())
```

```python
vectors.shape
```

```python
# now since all the movies are represented into vector representation , now we calculate cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# similarity of  first movie with each other movies
print(similarity[0])
# similarity of second movie with other movies
print(similarity[1])
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

**collaborative filtering**

**we need to search those users who have voted atleast 200 books and  select those books which has atleast 50 ratings**

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID', values = 'Book-Rating')
pt
```

```python
pt.fillna(0 , inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  # index_fetch
  index = np.where(pt.index == book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x: x[1] , reverse = True )[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```

```python
recommend('1984')
```

```python
recommend('Animal Farm')
```

**Recommendation using KNN algorithm**

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(pt)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
def recommend_book(book_name):
  book_id = np.where(pt.index == book_name)[0][0]
  distances, suggestions = model.kneighbors(pt.iloc[book_id,:].values.reshape(1,-1), n_neighbors = 6)

  for i in range(len(suggestions)):
    if i==0:
      print("The suggestion for" , book_name,"are :")
    if not i :
      print(pt.index[suggestions[i]])
```
